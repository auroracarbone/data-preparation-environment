```python
import pandas as pd
import numpy as np
```

```python
ghg = pd.read_csv('./input/emissioni_gas_serra.tsv', sep ='\t')
co2 = pd.read_csv('./input/co2-emission.tsv', sep = '\t')
rec = pd.read_csv('./input/recycling.tsv', sep = '\t')
```

```python
#voglio analizzare il dato di ciascun dataset corrispondente all'italia, quindi una specifica riga
ghg_ita = ghg.iloc[19,:].values
co2_ita = co2.iloc[19,:].values
rec_ita = rec.iloc[20,:].values
```

```python
ghg_ita_parsed = []

for i in range(len(ghg_ita)):
    if(isinstance(ghg_ita[i], np.float64)):
        ghg_ita_parsed.append(ghg_ita[i])
        continue 
    #la sintassi "continue" dice di proseguire con i cicli for --- in questo modo evito l'errore np.float64
    k = ghg_ita[i].split(' ')
    #la funzione split divide la stringa secondo un separatore che indichiamo, che in questo caso è lo spazio
    ghg_ita_parsed.append(k[0])
    
ghg_ita = np.array(ghg_ita_parsed)

co2_ita_parsed = []

for i in range(len(co2_ita)):
    if(isinstance(co2_ita[i], np.float64)):
        co2_ita_parsed.append(co2_ita[i])
        continue 
    #la sintassi "continue" dice di proseguire con i cicli for --- in questo modo evito l'errore np.float64
    k = co2_ita[i].split(' ')
    #la funzione split divide la stringa secondo un separatore che indichiamo, che in questo caso è lo spazio
    co2_ita_parsed.append(k[0])
    
co2_ita = np.array(co2_ita_parsed)

rec_ita_parsed = []

for i in range(len(rec_ita)):
    if(isinstance(rec_ita[i], np.float64)):
        rec_ita_parsed.append(rec_ita[i])
        continue 
    #la sintassi "continue" dice di proseguire con i cicli for --- in questo modo evito l'errore np.float64
    k = rec_ita[i].split(' ')
    #la funzione split divide la stringa secondo un separatore che indichiamo, che in questo caso è lo spazio
    rec_ita_parsed.append(k[0])
    
rec_ita = np.array(rec_ita_parsed)
```

```python
ita = pd.DataFrame({"ghg": ghg_ita, "co2": co2_ita, "recrate": rec_ita})
ita
#vedo che devo ripulire i dati
```

```python
#rimuovo riga intestazione
ita.drop(0, inplace = True)
```

```python
ita
```

```python
#converto tutto in float
ita['ghg'] = ita.ghg.astype(float)
ita['co2'] = ita.co2.astype(float)
ita['recrate'] = ita.recrate.astype(float)
```

```python
#mi resta da indicizzare il fatto che le righe 1 a 20 non sono valori casuali da 1 a 20, ma esprimono i valori dall'anno
#2000 al 2020
ita.index = np.arange(2000,2020,1)
```
