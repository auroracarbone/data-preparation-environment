```python
import pandas as pd
import numpy as np
```

```python
co2 = pd.read_csv('./input/co2-emission.tsv', sep = '\t') 
#dice che file leggere e che separatore viene utilizzato
#uso \t per il separatore perché aprendo il file vedevo che ciascun dato era separato da un "tab"
#ghg è la variabile con cui ho rinominato i dati raccolti da questo csv 
```

```python
co2.head()
#legge le prime 5 righe della tabella
```

```python
#vedo che la prima colonna contiene le nazioni, poi ho gli anni 
#la colonna 0,1,2,3,4 è aggiunta da pandas, non è davvero la colonna 0 della tabella, che è invece AT, BE, BG, ecc.
```

```python
countries = co2.iloc[:,0].values
#prendo tutte le righe, colonna 0 della tabella 
```

```python
countries
```

```python
#EXTRA: ho importato male i dati per qualche motivo, quindi ripeto lo stesso passaggio anche su countries per avere quello che hanno loro
countries_parsed = []

for i in range(len(countries)):
    k = countries[i].split(',')
    #la funzione split divide la stringa secondo un separatore che indichiamo, che in questo caso è lo spazio
    countries_parsed.append(k[1])
    
countries = countries_parsed
countries
```

```python
co2_19 = co2.iloc[:,-1].values
#prendo tutte le righe, colonna del 2019 della tabella 
```

```python
co2_19
```

```python
#se inserisco in un array l'ultima colonna osservo che non ho tutti numeri, ma alcuni contengono degli spazi e altri addirittura delle stringhe
#vogliamo vedere come renderli tutti numeri
```

```python
co2_19_parsed = []

for i in range(len(co2_19)):
    k = co2_19[i].split(' ')
    #la funzione split divide la stringa secondo un separatore che indichiamo, che in questo caso è lo spazio
    co2_19_parsed.append(k[0])
    
co2_19_parsed
```

```python
#traduco il valore in un array di numpy
co2_19 = np.array(co2_19_parsed)
co2_19
```

```python
co2_latest = pd.DataFrame({"country":countries, "avg_co2":co2_19, })
co2_latest
```

```python
co2_latest.drop(10,inplace=True)
co2_latest.drop(11,inplace=True)
co2_latest.drop(12,inplace=True)

co2_latest
```

```python
#ma questi non sono ancora numeri per python, perché devono perdere gli apici' ' . Vediamo come fare
#co2_latest.iloc[:,1] = co2_latest.iloc[:,1].astype(float)
#co2_latest.iloc[:,1] #ora ho un array di soli numeri 
#con questo metodo volendo, quando vado a fare la funzione sort, ottengo qualcosa di ordinato in modo crescente effettivamente
#nel tutorial vogliono eseguire la trasformazione della stringa in array in modo diverso (vedi passaggi dopo)
```

```python
co2_latest = co2_latest.sort_values(by='avg_co2')
co2_latest
#li ordina in modo crescente quasi tutti tranne gli ultimi due, perché non ho convertito i caratteri in numeri 
```

```python
co2_latest['avg_co2'] = co2_latest.avg_co2.astype(float)
co2_latest = co2_latest.sort_values(by='avg_co2')
co2_latest
#così è ordinato
```

```python
co2_latest.to_csv('./output/co2.csv',index=False)
```
